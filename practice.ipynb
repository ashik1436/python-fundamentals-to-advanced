```python
def print_digits(n):
    while n > 0:
        digit = n % 10
        print(digit)
        n = n // 10

print_digits(312)
```

```python
def count_digits(n):
  count=0
  while n>0:
    count+=1
    n=n//10
  return count
print(count_digits(312))
```

```python
def sum_of(n):
  total=0
  while n>0:
    total+=n%10
    n=n//10
  return total
print(sum_of(389))
```

```python
for i in range(1,101):
  if i%3 ==0 and i%5==0:
    print(i)
```

```python
while True:
    user_input = input("Enter number (or Quit): ")

    if user_input.lower() == "quit":
        break

    num = int(user_input)

    if num > 0:
        print("Positive")
    elif num < 0:
        print("Negative")
    else:
        print("Zero")
```

```python
def calculator(a,b,operation):
  if operation == '+':
    return a+b
  elif operation == '-':
    return a-b
  elif operation == '*':
    return a*b
  elif operation == '/':
    return a/b
  else:
    return "Invalid operation"
print(calculator(10,5,'+'))
```

```python
def is_prime(n):
  if n<2:
    return False
  for i in range(2,n):
    if n%i==0:
      return False
  return True
print(is_prime(7))
```

```python
secret = 7

while True:
    guess = int(input("Guess the number: "))

    if guess > secret:
        print("Too high")
    elif guess < secret:
        print("Too low")
    else:
        print("Correct!")
        break
```
